```python
import matplotlib.pyplot as plt

# Data
data = {'angry': {'orig': 1.55, '0': 1.15, '1': 1.2, '2': 1.3}, 'afraid': {'orig': 3.7, '0': 3.7, '1': 3.8, '2': 3.15}, 'happy': {'orig': 4.85, '0': 4.8, '1': 4.75, '2': 4.9}, 'sad': {'orig': 3.55, '0': 3.45, '1': 3.2, '2': 3.2}, 'surprised': {'orig': 3.9, '0': 3.4, '1': 3.45, '2': 3.2}, 'disgusted': {'orig': 4.05, '0': 3.7, '1': 3.7, '2': 3.6}}

# Convert data for plotting
x_labels = ['orig', '0', '1', '2']
series = {emotion: [values[x] for x in x_labels] for emotion, values in data.items()}

line_styles = ['-', '--', '-.', ':']
markers = ['o', 'v', '^', '<', '>', 's']
# Adjusting the plot to reflect the maximum score limitation

plt.figure(figsize=(10, 6))

for i, (emotion, values) in enumerate(series.items()):
    plt.plot(x_labels, values, label=emotion, marker=markers[i % len(markers)], linestyle=line_styles[i % len(line_styles)])

plt.title('Emotion Scores Over Iterations; 20 examples for each emotion')
plt.xlabel('Iteration')
plt.ylabel('Scores')
plt.ylim(0, 5)  # Set the y-axis to show scores from 0 to 5
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data = {
    'afraid': {'orig': 3.7, '0': 3.5, '1': 3.6, '2': 3.25},
}

# Convert data for plotting
x_labels = ['orig', '0', '1', '2']
series = {emotion: [values[x] for x in x_labels] for emotion, values in data.items()}

line_styles = ['-', '--', '-.', ':']
markers = ['o', 'v', '^', '<', '>', 's']
# Adjusting the plot to reflect the maximum score limitation

plt.figure(figsize=(10, 6))

for i, (emotion, values) in enumerate(series.items()):
    plt.plot(x_labels, values, label=emotion, marker=markers[i % len(markers)], linestyle=line_styles[i % len(line_styles)])

plt.axhline(y=3.65, color='r', linestyle='--', label='prefix control w/ KL; epoch 6')
plt.axhline(y=3.6, color='y', linestyle='--', label='prefix control w/o KL; epoch 6')
plt.axhline(y=3.3, color='g', linestyle='--', label='Baseline (System Prompt)')
plt.title('Emotion Scores Over Iterations; 20 examples for each emotion')
plt.xlabel('Iteration')
plt.ylabel('Scores')
plt.ylim(0, 5)  # Set the y-axis to show scores from 0 to 5
plt.legend()
plt.grid(True)
plt.show()
```
