```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('./suffix_gradient/merged_eval_results/exploited/test.csv')
```

```python
# Find the maximum value in each row
max_values = df.max(axis=1)

# Find the column index (name) of the maximum value in each row
argmax_columns = df.idxmax(axis=1)

# Print or store the results
print("Maximum values in each row:\n", max_values)
print("Column indices (names) of maximum values in each row:\n", argmax_columns)
```

```python
result_df = pd.DataFrame({
    'MaxValue': max_values,
    'ArgMaxColumn': argmax_columns
})

print(result_df)
```

```python
result_df.to_csv("attack_setting.csv")
```
